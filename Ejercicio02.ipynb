```python
import math

# Calcular pi  4 * suma de la serie de arctan(1)
def aproximar_pi(n):
    suma = 0.0
    for i in range(n):
        suma += ((-1) ** i) / (2 * i + 1)
    return 4 * suma

# Inciso (a): usar exactamente 5 trminos
n_a = 5
pi_a = aproximar_pi(n_a)
error_a = abs(pi_a - math.pi)

# Inciso (b): usar 500 trminos, como en el PDF
n_b = 500
pi_b = aproximar_pi(n_b)
error_b = abs(pi_b - math.pi)

# Imprimir resultados
print("Inciso (a):")
print(f"  n = {n_a}, pi  {pi_a:.6f}, error = {error_a:.2e}")

print("Inciso (b):")
print(f"  n = {n_b}, pi  {pi_b:.12f}, error = {error_b:.2e}")



```
