```python
import math
import cmath  

def resolver_cuadratica(a, b, c):
    D = b**2 - 4*a*c
    
    if D >= 0:
        sqrt_D = math.sqrt(D)
        if b > 0:
            x1 = (-b - sqrt_D) / (2 * a)
            x2 = (2 * c) / (-b - sqrt_D)
        else:
            x1 = (-b + sqrt_D) / (2 * a)
            x2 = (2 * c) / (-b + sqrt_D)
    else:
        sqrt_D = cmath.sqrt(D)
        x1 = (-b + sqrt_D) / (2 * a)
        x2 = (-b - sqrt_D) / (2 * a)
    
    return x1, x2

a, b, c = 1, -3, 2
x1, x2 = resolver_cuadratica(a, b, c)
print(f"Races: x1 = {x1}, x2 = {x2}")
```
