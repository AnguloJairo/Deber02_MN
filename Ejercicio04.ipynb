```python
def algoritmo_a(valores):
    product = 0
    for x in valores:
        product *= x
    return product

def algoritmo_b(valores):
    product = 1
    for x in valores:
        product *= x
    return product

def algoritmo_c(valores):
    product = 1
    for x in valores:
        if x == 0:
            return 0
        product *= x
    return product

# Prueba con un conjunto que incluye 0
valores = [2, 3, 0, 5]

print("Algoritmo a:", algoritmo_a(valores),"Incorrecto: siempre da 0")  # Incorrecto: siempre da 0
print("Algoritmo b:", algoritmo_b(valores), "Correcto: da 0 por el cero")  # Correcto: da 0 por el cero
print("Algoritmo c:", algoritmo_c(valores),"Correcto y más eficiente: da 0")  # Correcto y más eficiente: da 0
```
