```python
import math

# Calcular arctan(x) por serie de Maclaurin con n términos
def arctan(x, n):
    suma = 0.0
    for i in range(n):
        termino = ((-1)**i) * (x**(2*i + 1)) / (2*i + 1)
        suma += termino
    return suma

# Calcular pi usando fórmula de Machin con n términos
def aproximar_pi_machin(n):
    arctan_1_5 = arctan(1/5, n)
    arctan_1_239 = arctan(1/239, n)
    pi_aprox = 4 * (4 * arctan_1_5 - arctan_1_239)
    return pi_aprox



```

## Buscar el mínimo n tal que el error sea menor a 10^-3

```python
# Buscar el mínimo n tal que el error sea menor a 10^-3
def encontrar_n_machin(tolerancia):
    n = 1
    while True:
        pi_aprox = aproximar_pi_machin(n)
        error = abs(pi_aprox - math.pi)
        if error < tolerance:
            return n, pi_aprox, error
        n += 1
```

```python
tolerance = 1e-3
n_machin, pi_machin, error_machin = encontrar_n_machin(tolerance)

# Mostrar resultado
print(f"Número de términos necesarios: {n_machin}")
print(f"π aproximado: {pi_machin:.6f}")
print(f"Error: {error_machin:.2e}")
```
