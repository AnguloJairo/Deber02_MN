```python
import math

def truncar_tres_cifras(x):
    if x == 0:
        return 0.0
    else:
        orden = math.floor(math.log10(abs(x)))
        factor = 10**(2 - orden)
        return math.trunc(x * factor) / factor


```

## Suma hacia adelante (i = 1 a 10)

```python
def suma_hacia_adelante():
    suma = 0.0
    for i in range(1, 11):
        termino = 1 / (i ** 2)
        termino_truncado = truncar_tres_cifras(termino)
        suma += termino_truncado
    return round(suma, 3)
```

## Suma hacia atrás (i = 10 a 1)

```python
def suma_hacia_atras():
    suma = 0.0
    for i in range(10, 0, -1):
        termino = 1 / (i ** 2)
        termino_truncado = truncar_tres_cifras(termino)
        suma += termino_truncado
    return round(suma, 3)
```

```python

adelante = suma_hacia_adelante()
atras = suma_hacia_atras()

print(f"Suma hacia adelante: {adelante}")
print(f"Suma hacia atrás: {atras}")
```
