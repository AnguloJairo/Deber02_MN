### Suma Directa

```python
def suma_directa(a, b):
    n = len(a)
    suma = 0
    for i in range(n):
        for j in range(i + 1):
            suma += a[i] * b[j]
    return suma

```

### Suma Optimizada

```python
def suma_optimizada(a, b):
    n = len(a)
    B = [0] * n
    B[0] = b[0]
    for i in range(1, n):
        B[i] = B[i-1] + b[i]

    suma = 0
    for i in range(n):
        suma += a[i] * B[i]
    return suma
```

```python
# Ejemplo de uso
a = [1, 2, 3, 4]
b = [1, 1, 1, 1]

print("Suma directa:", suma_directa(a, b))
print("Suma optimizada:", suma_optimizada(a, b))
```
